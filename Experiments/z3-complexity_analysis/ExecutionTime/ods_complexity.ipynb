```python
import pandas as pd, numpy as np, json, time

from ods import Sample, ODS
np.seterr(invalid='ignore')
```

```python
%%time
df = pd.read_csv('extended_bgp_testbed_5.csv.gz')
df = df.drop(['Unnamed: 0'], axis=1)
df = df.dropna()
df = df.astype('float64')
df.shape
```

```python
samples_to_use_list = list(range(1000, 10000, 1000))
samples_to_use_list = samples_to_use_list + list(range(10000, 110000, 10000))
```

```python
results = {}
results['algorithm'] = 'dbscan'
results['times'] = []

for sample_to_use in samples_to_use_list:
            
    print('ODS: {}'.format(sample_to_use))
    results_samples = []
    dfNormalized = df[:sample_to_use]
        
    num_runs = 10        
        
    for measurement in range(num_runs):
        sampleSkip = 30
        bufferDF = dfNormalized[0:sampleSkip]
        testDF = dfNormalized[sampleSkip:]

        ods = ODS(lamb = 0.125,\
                    epsilon = 'dynamic',\
                    beta = 0.4,\
                    mu = 'auto',\
                    startingBuffer = bufferDF.values,
                    tp = 'auto')
        ods.runInitialization()    

        samples = testDF.values
        samples_to_samples = []
        for sample in samples:
            samples_to_samples.append(Sample(sample,0))

        start = time.time()
        for sample in samples_to_samples:
            result = ods.runOnNewSample(sample)
        end = time.time()
        time_interval = end-start
        results_samples.append(time_interval)
        print('{} - {}'.format(sample_to_use, time_interval))
    
    results['times'].append(results_samples)
    with open('Results/ods_extended.json', 'w') as f:
        json.dump(results, f, indent=2)    
```
