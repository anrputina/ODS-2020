```python
import pandas as pd, numpy as np, json, time
from sklearn.cluster import DBSCAN
```

```python
%%time
df = pd.read_csv('extended_bgp_testbed_5.csv.gz').drop(['Unnamed: 0'], axis=1)
df = df.dropna()
df = df.astype('float64')
df.shape
```

```python
samples_to_use_list = list(range(1000, 10000, 1000))
samples_to_use_list = samples_to_use_list + list(range(10000, 110000, 10000))
```

```python
results = {}
results['algorithm'] = 'dbscan'
results['times'] = []

for index, sample_to_use in enumerate(samples_to_use_list[:15]):
    
    print('DBScan: {}'.format(sample_to_use))
    
    results_samples = []
    
    if index <= 9:
        num_runs = 10
    else:
        num_runs = 4
    
    for measurement in range(num_runs):
    
        start = time.time()
        dbscan = DBSCAN(eps=6,
                        min_samples=18)

        dbscan.fit(df[:sample_to_use])
        end = time.time()
        time_interval = end - start
    
        results_samples.append(time_interval)
        print('{} - {}'.format(sample_to_use, time_interval))
        
    results['times'].append(results_samples)
    with open('Results/dbscan_extended.json', 'w') as f:
        json.dump(results, f, indent=2)    
```
